# RAVDESS (Ryerson Audio-Visual Database of Emotional Speech and Song)

Il Ryerson Audio-Visual Database of Emotional Speech and Song (RAVDESS) contiene 7.356 file (dimensione totale: 24,8 GB). Il database contiene 24 attori professionisti (12 femmine, 12 maschi), che vocalizzano due affermazioni lessicalmente abbinate con un accento nordamericano neutro. Il discorso(Speech) include espressioni calme, felici, tristi, arrabbiate, paurose, sorprese e disgustate, e il canto(Song) contiene emozioni calme, felici, tristi, arrabbiate e paurose. Ogni espressione è prodotta a due livelli di intensità emotiva (normale, forte), con un'espressione neutra aggiuntiva. Tutte le condizioni sono disponibili in tre formati di modalità: solo audio (16 bit, 48kHz .wav), audio-video (720p H.264, AAC 48kHz, .mp4) e solo video (senza audio). Nota, non ci sono file di canzoni per Actor_18.

## Convenzione di denominazione dei file
Il file vocale (Audio_Speech_Actors_01-24.zip, 215 MB) contiene 1440 file: 60 prove per attore x 24 attori = 1440.<br>
Ciascuno dei file RAVDESS ha un nome file univoco. Il nome del file è costituito da un identificatore numerico in 7 parti (ad esempio, 02-01-06-01-02-01-12.mp4).

Filename identifiers:

- Modality (01 = full-AV, 02 = video-only, 03 = audio-only).
- Vocal channel (01 = speech, 02 = song).
- Emotion (01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised).
- Emotional intensity (01 = normal, 02 = strong). NOTE: There is no strong intensity for the 'neutral' emotion.
- Statement (01 = "Kids are talking by the door", 02 = "Dogs are sitting by the door").
- Repetition (01 = 1st repetition, 02 = 2nd repetition).
- Actor (01 to 24. Odd numbered actors are male, even numbered actors are female).

# Creazione file csv
Prima di creare il file csv, dobbiamo cambiare il cirectory con quello che punta al set di dati del padre di ravdess (Audio_Speech_Actors_01-24).

```python
from os.path import dirname
ravdess = new_path # or os.getcwd()
path = dirname(ravdess)
os.chdir(path)
```

```python
os.getcwd()
```

```python
dirs = os.listdir(ravdess)
file_emotion = []
file_emotional_intensity = []
file_statement = []
file_path = []
for dir in dirs:
    actor = os.listdir(ravdess + '/' + dir)
    for audio in actor:
        part = audio.split('.')[0]
        part = part.split('-')
        file_emotion.append(int(part[2]))
        file_emotional_intensity.append(int(part[3]))
        file_statement.append(int(part[4]))
        file_path.append(ravdess + '/' + dir + '/' + audio)
```

## Creazione di un dataframe pandas con le emozioni e il percorso audio

```python
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])
emotional_intensity_df = pd.DataFrame(file_emotional_intensity, columns=['Emotional_intensity'])
statement_df = pd.DataFrame(file_statement, columns=['Statement'])
path_df = pd.DataFrame(file_path, columns=['Path'])
ravdess_df = pd.concat([path_df, emotion_df, emotional_intensity_df, statement_df], axis=1)
```

```python
ravdess_df.Emotions.replace({1:'neutral', 2:'calm', 3:'happy', 4:'sad', 5:'angry', 6:'fearful', 7:'disgust', 8:'surprised'}, inplace=True)
ravdess_df.Emotional_intensity.replace({1:'normal', 2:'strong'}, inplace=True)
ravdess_df.Statement.replace({1:'Kids are talking by the door', 2:'Dogs are sitting by the door'}, inplace=True)
ravdess_df.head()
```

```python
dataset = pd.concat([ravdess_df], axis=0)
dataset.to_csv("dataset.csv", index=False)
dataset.tail()
```

# Estrazione feature con librosa

- __chroma_stft(*[, y, sr, S, norm, n_fft, ...])__ Calcola un cromagramma da una forma d'onda o uno spettrogramma di potenza.

- __mfcc(*[, y, sr, S, n_mfcc, dct_type, norm, ...])__ Coefficienti cepstrali di frequenza Mel (MFCC)

- __tonnetz(*[, y, sr, chroma])__ Calcola le caratteristiche del centroide tonale (tonnetz)

## Spectrogram

A spectrogram is a visual way of representing the signal strength, or “loudness”, of a signal over time at various frequencies present in a particular waveform. Not only can one see whether there is more or less energy at, for example, 2 Hz vs 10 Hz, but one can also see how energy levels vary over time.<br>
A spectrogram is usually depicted as a heat map, i.e., as an image with the intensity shown by varying the color or brightness.<br>
We can display a spectrogram using. librosa.display.specshow:

    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()

.stft() converts data into short term Fourier transform. STFT converts signals such that we can know the amplitude of the given frequency at a given time. Using STFT we can determine the amplitude of various frequencies playing at a given time of an audio signal. .specshow is used to display a spectrogram.<br>
The vertical axis shows frequencies (from 0 to 10kHz), and the horizontal axis shows the time of the clip. Since we see that all action is taking place at the bottom of the spectrum, we can convert the frequency axis to a logarithmic one.

    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar()

```python
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db, x_axis='time',y_axis='log', ax=ax)
fig.colorbar(img, ax=ax, format=f'%0.2f')
```
