# RAVDESS (Ryerson Audio-Visual Database of Emotional Speech and Song)

Il Ryerson Audio-Visual Database of Emotional Speech and Song (RAVDESS) contiene 7.356 file (dimensione totale: 24,8 GB). Il database contiene 24 attori professionisti (12 femmine, 12 maschi), che vocalizzano due affermazioni lessicalmente abbinate con un accento nordamericano neutro. Il discorso(Speech) include espressioni calme, felici, tristi, arrabbiate, paurose, sorprese e disgustate, e il canto(Song) contiene emozioni calme, felici, tristi, arrabbiate e paurose. Ogni espressione è prodotta a due livelli di intensità emotiva (normale, forte), con un'espressione neutra aggiuntiva. Tutte le condizioni sono disponibili in tre formati di modalità: solo audio (16 bit, 48kHz .wav), audio-video (720p H.264, AAC 48kHz, .mp4) e solo video (senza audio). Nota, non ci sono file di canzoni per Actor_18.

## Convenzione di denominazione dei file
Il file vocale (Audio_Speech_Actors_01-24.zip, 215 MB) contiene 1440 file: 60 prove per attore x 24 attori = 1440.<br>
Ciascuno dei file RAVDESS ha un nome file univoco. Il nome del file è costituito da un identificatore numerico in 7 parti (ad esempio, 02-01-06-01-02-01-12.mp4).

Filename identifiers:

- Modality (01 = full-AV, 02 = video-only, 03 = audio-only).
- Vocal channel (01 = speech, 02 = song).
- Emotion (01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised).
- Emotional intensity (01 = normal, 02 = strong). NOTE: There is no strong intensity for the 'neutral' emotion.
- Statement (01 = "Kids are talking by the door", 02 = "Dogs are sitting by the door").
- Repetition (01 = 1st repetition, 02 = 2nd repetition).
- Actor (01 to 24. Odd numbered actors are male, even numbered actors are female).

# Creazione file csv
Prima di creare il file csv, dobbiamo cambiare il cirectory con quello che punta al set di dati del padre di ravdess (Audio_Speech_Actors_01-24).

```python
dirs = os.listdir(ravdess)
file_emotion = []
file_emotional_intensity = []
file_statement = []
file_actor = []
file_actor_gender = []
file_path = []
for dir in dirs:
    actor = os.listdir(ravdess + '/' + dir)
    for audio in actor:
        part = audio.split('.')[0]
        part = part.split('-')
        file_emotion.append(int(part[2]))
        file_emotional_intensity.append(int(part[3]))
        file_statement.append(int(part[4]))
        file_actor.append(int(part[6]))
        file_actor_gender.append(int(part[6])%2)
        file_path.append(ravdess + '/' + dir + '/' + audio)
```

# Estrazione feature con librosa

- __chroma_stft(*[, y, sr, S, norm, n_fft, ...])__ Calcola un cromagramma da una forma d'onda o uno spettrogramma di potenza.

- __melspectrogram(*[, y, sr, S, n_fft, ...])__ Calcola uno spettrogramma in scala mel.

- __mfcc(*[, y, sr, S, n_mfcc, dct_type, norm, ...])__ Coefficienti cepstrali di frequenza Mel (MFCC)

- __spectral_contrast(*[, y, sr, S, n_fft, ...])__ Calcola il contrasto spettrale

- __tonnetz(*[, y, sr, chroma])__ Calcola le caratteristiche del centroide tonale (tonnetz)

```python
chroma_stft = []
melspectrogram = []
mfcc = []
spectral_contrast = []
tonnetz = []


for audio_file in ravdess_df["Path"]:
    data, sample_rate = librosa.load(audio_file)
    
    # Chroma_stft
    stft = np.abs(librosa.stft(data))
    chroma_stft.append(np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T, axis=0))

    # MelSpectogram
    melspectrogram.append(np.mean(librosa.feature.melspectrogram(y=data, sr=sample_rate).T, axis=0))
    
    # MFCC
    mfcc.append(np.mean(librosa.feature.mfcc(y=data, sr=sample_rate).T, axis=0))
    
    # Spectral_contrast
    spectral_contrast.append(np.mean(librosa.feature.spectral_contrast(y=data, sr=sample_rate).T, axis=0))

    # Tonnetz
    tonnetz.append(np.mean(librosa.feature.tonnetz(y=data).T, axis=0))
```

Creo i dataframe: chroma 12, mel 128, mfcc 20, contrast 7, tonnetz 6 colonne.

Il metodo rename di default non cambia il valore delle colonne del dataframe (inplace vale false di default) ma restituisce un nuovo dataframe con le colonne cambiate. Devi quindi assegnarlo alla variabile

```python
mel_df = pd.DataFrame(melspectrogram)
mel_df.rename(columns=lambda i: 'MelSpectrogram_' + str(i+1), inplace=True)
```

```python
mfcc_df = pd.DataFrame(mfcc)
mfcc_df.rename(columns=lambda i: 'MFCC_' + str(i+1), inplace=True)
```

```python
contrast_df = pd.DataFrame(spectral_contrast)
contrast_df.rename(columns=lambda i: 'SpectralContrast_' + str(i+1), inplace=True)
```

```python
tonnetz_df = pd.DataFrame(tonnetz)
tonnetz_df.rename(columns=lambda i: 'Tonnetz_' + str(i+1), inplace=True)
```

Funziona anche in modo molto efficace nel rumore; il sistema uditivo umano non è lineare; pertanto, MFCC utilizza la scala Mel;
- **SpectralContrast_x**: viene stimata l'energia media di un suono e il valore di contrasto elevato è evidente e il valore di contrasto basso corrisponde al rumore a banda larga;
- **Tonnetz_x**: le caratteristiche di Tonnetz funzionano allo stesso modo come il cromatogramma, è anche correlato all'armonia e alla classe del tono.

## Cross validation

```python
training_actors = training_set['Actor']
training_actors = training_actors.unique()
```

```python
training_actors_m = []
training_actors_f = []
for actor in training_actors:
    if (actor%2 == 0):
        training_actors_f.append(actor)
    else:
        training_actors_m.append(actor)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
classifier = RandomForestClassifier()
```

Cross-fold validation con 10 fold (ciascuono composto da 2 attori, 1 maschio e 1 femmina)

```python
accuracy = []
f1 = []
for i in range(10):
    m = training_actors_m[i]
    f = training_actors_f[i]
        
    cross_test_set = training_set.loc[training_set['Actor'].isin([m, f])]
    cross_training_set = training_set.drop(cross_test_set.index)
       
    y_train = cross_training_set["Emotions"]
    X2d_train = cross_training_set[dataset.columns[6:]]
    y_test = cross_test_set["Emotions"]
    X2d_test = cross_test_set[dataset.columns[6:]]
    
    # fit the classifier
    classifier.fit(X2d_train, y_train)
    # compute the score and record it
    accuracy.append(classifier.score(X2d_test, y_test))
    y_pred = classifier.predict(X2d_test)
    f1.append(f1_score(y_test, y_pred, average="macro"));
```

```python
# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Pandas dataframe
stats = pd.DataFrame({"Accuracy": accuracy, "F1-Score": f1})

# Plot the dataframe
ax = stats[['Accuracy', 'F1-Score']].plot(kind='box', title='Cross-Validation stats')

# Display the plot
plt.show()
```
